## Commissioning Instrument Data

```python
def loadPM(name='/Data/CI/20190524/00013116/guide-00013116.fits.fz'):
    images = {}
    hdus = fitsio.FITS(name)
    for camera in ('CIN', 'CIE', 'CIS', 'CIW', 'CIC'):
        images[camera] = hdus[camera][0, :, :][0]
    return images
```
